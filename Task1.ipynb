```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("titanic_dataset.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(columns=["Age"])
dataset.drop(columns=["Cabin"])
```

```python
gender_counts = dataset['Sex'].value_counts()
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
dataset['Age'].dropna().plot(kind='hist', bins=30, edgecolor='black')
plt.title('Age Distribution on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
